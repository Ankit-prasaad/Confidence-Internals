```python
import numpy as np
from scipy.stats import t, norm
import math
```

```python
# Data
data = [1.13, 1.55, 1.43, 0.92, 1.25,
        1.36, 1.32, 0.85, 1.07, 1.48,
        1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
data
```

```python
# Sample statistics
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)
print("Sample Mean:", mean)
print("Sample Standard Deviation:", std)
```

```python
#  t confidence interval
confidence = 0.99
alpha = 1 - confidence
t_critical = t.ppf(1 - alpha/2, df=n-1)

margin_error_t = t_critical * (std / math.sqrt(n))

lower_t = mean - margin_error_t
upper_t = mean + margin_error_t

print("\\n99% CI using t-distribution:")
print((lower_t, upper_t))
```

```python
# z confidence interval
sigma = 0.2
z_critical = norm.ppf(1 - alpha/2)

margin_error_z = z_critical * (sigma / math.sqrt(n))

lower_z = mean - margin_error_z
upper_z = mean + margin_error_z

print("\\n99% CI using z-distribution:")
print((lower_z, upper_z))
```
